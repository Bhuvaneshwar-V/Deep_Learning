# Remove the unwanted column

```python
df = df.drop(columns = ['id', 'Unnamed: 32'], axis = 1)
```

# Convert the string column to numerical column

```python
df['diagnosis'] = df['diagnosis'].map({'M' : 1, 'B' : 0})
```

# Visualizing the target class

```python
df['diagnosis'].value_counts().plot(kind = 'pie')
plt.title('Class Distribution')
```

# Store the independent feature in 'x' & dependent feature in 'y'

```python
x = df.drop(columns = ['diagnosis'], axis =1)
y = df['diagnosis']
```

# Train/Test split
- Split data into two part : a training set and a testing set
- Train the model(s) on training set
- Test the Model(s) on Testing set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 42)
```

# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train = scaler.fit_transform(x_train)
X_test = scaler.transform(x_test)
```

# Load the Model

```python
import tensorflow
from tensorflow import keras
from sklearn.metrics import accuracy_score
```

# Create the Neural Network

```python
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(x_train.shape[1],)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(1, activation='sigmoid')
])
```

# Compiling the Neural Network

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

# Training the Neural Network

```python
history = model.fit(X_train, y_train, validation_split=0.2, epochs=50)
```

# Visualizing the accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training accuracy', 'validation accuracy'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training loss', 'validation loss'], loc = 'upper right')
```

# Evaluation

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(accuracy)
```
